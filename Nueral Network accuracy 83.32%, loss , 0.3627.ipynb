```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
```

```python
df=pd.read_csv('fall_new.csv')
```

```python
df.head()
```

```python
X=df[['TIME','SL','EEG','BP','HR','CIRCLUATION']]
y=df['FALL']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y.values.reshape(-1,1), random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# One-hot encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
y_train_categorical 
```

```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense

model.add(Dense(units=6, activation='relu', input_dim=6))
model.add(Dense(units=2, activation='softmax'))
```

```python
number_classes = 2
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model = Sequential()
model.add(Dense(units=40, activation='linear', input_dim=6))
model.add(Dense(units=40, activation='tanh'))
model.add(Dense(units=40, activation='sigmoid'))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.compile(optimizer=tf.train.RMSPropOptimizer(0.01),
              loss=tf.keras.losses.categorical_crossentropy,
              metrics=[tf.keras.metrics.categorical_accuracy])
```

```python
model.compile(optimizer=tf.train.GradientDescentOptimizer(0.01),
              loss=tf.keras.losses.categorical_crossentropy,
              metrics=[tf.keras.metrics.categorical_accuracy])
```

```python
model.summary()
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Save the model
model.save("model_trained.h5")
```
